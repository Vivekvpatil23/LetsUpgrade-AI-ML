```python
import pandas as pd
import numpy as py
from sklearn.preprocessing import LabelEncoder
data=pd.read_csv("general_data.csv")
```

```python
data.columns
```

```python
data.isnull()
data.dropna()
data.drop_duplicates()
```

```python
data["Attrition"].replace(to_replace=("No","Yes"),value=(0,1),inplace=True)
data
```

```python
catg=[i for i in data.columns if data.dtypes[i]== 'object']
catg
```

```python
labenc=LabelEncoder()
data.Attrition=labenc.fit_transform(data.Attrition)
data['BusinessTravel']=labenc.fit_transform(data['BusinessTravel'])

data['Department']=labenc.fit_transform(data['Department'])

data['EducationField']=labenc.fit_transform(data['EducationField'])

data['Gender']=labenc.fit_transform(data['Gender'])

data['JobRole']=labenc.fit_transform(data['JobRole'])

data['MaritalStatus']=labenc.fit_transform(data['MaritalStatus'])
data
```

```python
df_yes=data[data['Attrition']==1]
df_No=data[data['Attrition']==0]
```

# ManWhitney Test

```python
from scipy.stats import mannwhitneyu
stats,p=mannwhitneyu(df_yes.Age,df_No.Age)
print(stats,p)
```

Here value of p = 2.9951588479067175e-30
p value is less than 0.05,so we reject the null value and select the alternate value

```python
from scipy.stats import mannwhitneyu
stats,p=mannwhitneyu(df_yes.BusinessTravel,df_No.BusinessTravel)
print(stats,p)
```

Here p = 0.024442388462694316
p value is less than 0.05,so we cannot accept the null value and select the alternate value

# Chi SquareTest

```python
from scipy.stats import chi2_contingency
chisqtable=pd.crosstab(data.Attrition,data.MaritalStatus)
chisqtable
```

```python
stats,p,dof,expected=chi2_contingency(chisqtable)
```

```python
print(stats,p)
```

Here p=8.45385940605786e-31
p value is less than 0.05, so cannot take the null value and select the alternate value.
